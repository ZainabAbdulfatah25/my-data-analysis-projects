```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the total number of fellows assigned to 4U TALENT ACCELERATOR
df = pd.read_csv('/')

# Check the first few rows of the dataframe
print(df.head())

# Data Cleaning

# Checking for missing values
print("\nMissing values:\n")
print(df.isnull().sum())

# Checking for duplicates
print("\nDuplicate found:\n")
print(df.duplicated().sum())

# Checking the size of the dataframe
print(f"\nDataFrame size: {df.size}")
print(f"DataFrame shape: {df.shape}")

# Create a list of the desired column names
Fellows_attendance_columns = ['Percentage of Attendance', 'Fullname', 'Fellow ID']

# Selecting the desired columns from df
Fellows_attendance = df[Fellows_attendance_columns]

# Display the selected columns
print("\nSelected Columns:\n")
print(Fellows_attendance.head())
df.value_counts(normalize=True)
# Common Attendance by Fullname
region_diagnosis = df.groupby(['Percentage of Attendance', 'Fullname']).size().head().unstack().fillna(0)
print("\nPercentage of Attendance by Fullname:\n", region_diagnosis)
#Demographics for Specific Conditions (e.g., Percentage of Attendance)
Fellows_attendance_columns = df[df['Percentage of Attendance'] >= 75][['Percentage of Attendance']]
print("\nPercentage of Attendance Demographics:\n", Fellows_attendance_columns)
# basic statistics for 'Percentage of Attendance' column
attendance_summary = df['Percentage of Attendance'].describe()
#summary
print("\nSummary Statistics for Percentage of Attendance:\n", attendance_summary)
```
