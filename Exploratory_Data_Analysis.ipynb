```python
# Exploratory Data Analysis (EDA)
# Sample Employee Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create Dataset
data = {
    'EmployeeID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Department': ['IT', 'HR', 'Sales', 'Finance', 'Marketing', 'IT', 'HR', 'Sales', 'Marketing', 'Finance'],
    'ExperienceYears': [5, 3, 8, 2, 6, 10, 1, 4, 9, 7],
    'MonthlySalary': [2500, 2000, 3000, 1800, 2700, 4000, 1500, 2200, 3500, 3200],
    'RemoteWork': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'PerformanceScore': [80, 70, 88, 65, 75, 90, 60, 72, 85, 78],
    'ProjectsCompleted': [6, 4, 7, 3, 5, 8, 2, 4, 7, 6]
}

df = pd.DataFrame(data)

print("✅ Dataset created successfully!\n")
print(df.head())

# Basic Info
print("\n📘 Dataset Info:")
df.info()

print("\n📊 Shape of dataset:", df.shape)
print("\n🔤 Column names:", df.columns.tolist())

# Check Missing Values
print("\n🧩 Missing Values per column:\n", df.isnull().sum())

plt.figure(figsize=(6, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='coolwarm')
plt.title("Missing Values Heatmap")
plt.xlabel("Columns")
plt.ylabel("Records")
plt.show()

# Descriptive Statistics
print("\n📈 Statistical Summary:")
print(df.describe())

# Univariate Analysis
# Numeric Histogram with Labels
plt.figure(figsize=(6, 4))
ax = sns.histplot(df['MonthlySalary'], kde=True, color='skyblue', bins=8)
plt.title("Monthly Salary Distribution")
plt.xlabel("Monthly Salary ($)")
plt.ylabel("Number of Employees")

# Add numeric labels on bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width()/2., height + 0.2, int(height), ha="center")

plt.show()

# Categorical Count Plot with Labels
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Department', data=df, palette='Set2')
plt.title("Department Distribution")
plt.xlabel("Department")
plt.ylabel("Number of Employees")

# Add number labels
for container in ax.containers:
    ax.bar_label(container, fmt='%d')

plt.show()

# Bivariate Analysis
# Numerical vs Numerical
plt.figure(figsize=(6, 4))
sns.scatterplot(x='ExperienceYears', y='MonthlySalary', data=df, hue='Department', s=100)
plt.title("Experience vs Monthly Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Monthly Salary ($)")
plt.legend(title='Department')
plt.show()

# Categorical vs Numerical (Boxplot)
plt.figure(figsize=(6, 4))
sns.boxplot(x='Department', y='MonthlySalary', data=df, palette='cool')
plt.title("Salary Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Monthly Salary ($)")
plt.show()

# Correlation Matrix
corr = df.corr(numeric_only=True)
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Between Numerical Features")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()

# Outlier Detection
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['MonthlySalary'], color='lightcoral')
plt.title("Outlier Detection in Monthly Salary")
plt.xlabel("Monthly Salary ($)")
plt.show()

Q1 = df['MonthlySalary'].quantile(0.25)
Q3 = df['MonthlySalary'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['MonthlySalary'] < (Q1 - 1.5 * IQR)) | (df['MonthlySalary'] > (Q3 + 1.5 * IQR))]
print("\n🚨 Number of outliers in MonthlySalary:", len(outliers))

# Group Analysis
avg_salary = df.groupby('Department')['MonthlySalary'].mean().sort_values(ascending=False)
print("\n💰 Average Salary by Department:\n", avg_salary)

plt.figure(figsize=(6, 4))
ax = sns.barplot(x=avg_salary.index, y=avg_salary.values, palette='viridis')
plt.title("Average Monthly Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary ($)")

# Add numeric labels on top of bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')

plt.show()

# Pairplot (Relationships)
sns.pairplot(df, diag_kind='kde', hue='Department')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()

# Insights Summary
print("\n📋 INSIGHTS SUMMARY")
print("- No missing values found.")
print("- Salary, experience, performance, and projects completed are positively correlated.")
print("- IT and Marketing departments show higher average salaries.")
print("- Minimal outliers; data is consistent and balanced.")
print("- Dataset is clean and ready for modeling or deeper analysis.")

# Automated EDA Tools
!pip install ydata-profiling sweetviz dtale

from ydata_profiling import ProfileReport
profile = ProfileReport(df, title="Pandas Profiling Report", explorative=True)
profile.to_file("EDA_Report.html")

import sweetviz as sv
report = sv.analyze(df)
report.show_html("Sweetviz_Report.html")

print("\n✅ EDA Complete — Dataset is ready for further analysis or modeling.")
```
