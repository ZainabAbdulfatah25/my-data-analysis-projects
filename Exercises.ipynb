```python
#Exercise 1
import pandas as pd

data = {
    "Product": ["Shirt", "Shoes", "Hat", "Bag", "Watch", "Shoes", "Hat", "Bag"],
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West"],
    "Sales": [25000, 35000, 150000, 200000, 35000, 97000, 82000, 22000],
    "Profit": [5000, 70000, 285000, 45000, 6000, 50000, 10000, 3000]
}

df = pd.DataFrame(data)
df
```

Tasks
1. Display the dataset information and summary statistics.
2. Find which region recorded the highest total sales.
3. Plot a bar chart comparing total sales per region.
Compute and visualize profit-to-sales ratio by product.


```python
#Exercise 1 Solution
# Task 1
print(df.info())
print(df.describe())

# Task 2
sales_by_region = df.groupby("Region")["Sales"].sum()
print(sales_by_region)

import matplotlib.pyplot as plt

#Task 3: Total Sales by Region
sales_by_region = df.groupby("Region")["Sales"].sum()
sales_by_region.plot(kind="bar", title="Total Sales by Region")
plt.ylabel("Sales")
plt.show()

# Task 4: Profit-to-Sales Ratio by Product as Percentage
df["Profit_Sales_Ratio"] = df["Profit"] / df["Sales"]
df["Profit_Sales_Percentage"] = df["Profit_Sales_Ratio"] * 100

ax = df.plot(
    x="Product",
    y="Profit_Sales_Percentage",
    kind="bar",
    color="green",
    legend=False
)

plt.title("Profit-to-Sales Ratio by Product")
plt.ylabel("Profit-to-Sales (%)")

# Set y-axis from 0 to 100%
ax.set_ylim(0, 100)
ax.set_yticks(range(0, 101, 10))  # interval of 10%
ax.set_yticklabels([f"{y}%" for y in range(0, 101, 10)])

plt.show()
```

```python
#Exercise 2
import pandas as pd

students = {
    "Student": ["Aisha", "Bola", "Chinedu", "Fatima", "John", "Kemi", "Tunde", "Zainab"],
    "Hours_Studied": [5, 8, 3, 9, 6, 4, 2, 7],
    "Exam_Score": [65, 88, 45, 92, 78, 60, 40, 85],
    "Attendance(%)": [80, 95, 70, 98, 85, 75, 60, 90]
}

df = pd.DataFrame(students)
df
```

Tasks
1. Display correlation between study hours and exam scores.
2. Identify students who scored above 80%.
3. Visualize relationship between attendance and exam performance.
4. Provide recommendations for improvement.


```python
import seaborn as sns
import matplotlib.pyplot as plt

#Task 1: Correlation#
corr_matrix = df[["Hours_Studied", "Exam_Score", "Attendance(%)"]].corr()
# Multiply correlations by 100 to express as percentage
corr_matrix_percent = corr_matrix * 100
print("Correlation Matrix (%):\n", corr_matrix_percent.round(2))

#Task 2: Top Students (Exam Score as % of max 100)#
top_students = df[df["Exam_Score"] > 80].copy()
top_students["Exam_Score(%)"] = top_students["Exam_Score"]
top_students["Attendance(%)"] = top_students["Attendance(%)"]
print("\nTop Students (%):\n", top_students[["Hours_Studied", "Exam_Score(%)", "Attendance(%)"]])

#Task 3: Scatterplot#
plt.figure(figsize=(8,6))
scatter = sns.scatterplot(
    data=df,
    x="Hours_Studied",
    y="Exam_Score",
    hue="Attendance(%)",
    palette="viridis",
    size="Attendance(%)",
    sizes=(50,200)
)

plt.xlabel("Hours Studied")
plt.ylabel("Exam Score (%)")
plt.title("Study Hours vs Exam Score (Attendance in %)")
plt.legend(title="Attendance (%)")
plt.show()
```

```python
#Exercise 3
import pandas as pd

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Total_Waste(kg)": [1200, 1150, 980, 870, 760, 700],
    "Recycled_Waste(kg)": [200, 250, 300, 350, 400, 450]
}

df = pd.DataFrame(data)
df
```

Tasks
1. Calculate the recycling percentage each month.
2. Visualize total vs recycled waste.
3. Determine the overall reduction rate.
4. Provide recommendations for sustainability programs.


```python
import matplotlib.pyplot as plt

# Recycling Rate (%)
df["Recycling_Rate(%)"] = (df["Recycled_Waste(kg)"] / df["Total_Waste(kg)"]) * 100

# figure and axis
fig, ax1 = plt.subplots(figsize=(10,6))

# Plot Total Waste and Recycled Waste on left y-axis
ax1.plot(df["Month"], df["Total_Waste(kg)"], label="Total Waste (kg)", marker='o')
ax1.plot(df["Month"], df["Recycled_Waste(kg)"], label="Recycled Waste (kg)", marker='o')
ax1.set_xlabel("Month")
ax1.set_ylabel("Waste (kg)")
ax1.tick_params(axis='y')
ax1.legend(loc="upper left")

#y-axis Recycling Rate (%)
ax2 = ax1.twinx()
ax2.plot(df["Month"], df["Recycling_Rate(%)"], label="Recycling Rate (%)", color="green", marker='s', linestyle='--')
ax2.set_ylabel("Recycling Rate (%)")
ax2.set_ylim(0, 100)
ax2.tick_params(axis='y')
ax2.legend(loc="upper right")

plt.title("Waste Management Trends and Recycling Rate (%)")
plt.show()
```
