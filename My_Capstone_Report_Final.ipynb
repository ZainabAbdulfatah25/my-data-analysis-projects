```python
# Data Cleaning
# Checking for missing values
print("Missing values:\n")
df.isnull().sum()
```

```python
#Replace NaN values in the 'Age' column with 'random numbers'
df['Age'] = df['Age'].fillna('random numbers')
df.dropna(inplace=True)
df.head(30)
```

```python
#Replace NaN values in the 'Diagnosis' column with 'random valuea'
df['Diagnosis'] = df['Diagnosis'].fillna('random values')
df.dropna(inplace=True)
df.head(30)
```

```python
# Data Cleaning
# Checking for missing values
print("Missing values:\n")
df.isnull().sum()
```

```python
#Checking for duplicates
print("\nDuplicates found:\n")
print(df.duplicated().sum())

# Duplicates found:
```

```python
crosstab_group = pd.crosstab(df.head()['PatientID'],df['Gender'],values=df['Date'],aggfunc='count')
# Print the cross-tabulation
print(crosstab_group)
df.head()
```
