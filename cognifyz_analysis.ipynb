**LEVEL 1 DATA PREPROCESSING**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/Dataset  (2).csv')
print(df.head(10))
df.info()
#handle missing values
print(df.isnull())
# drop missing values w
df.dropna(inplace=True)
print(df)
# Count total duplicates
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
df.size
df.shape
```

**TASK 1**

> **TOP CUISINE ANALYSIS**


* STEP1:  Analyze top 3 cuisines
* STEP2:  Calculate the percentage restaurants offering this cuisines
* STEP3:  visualize the results






```python
 #TOP CUISINES
 #Step 1:The occurrences of each cuisine
cuisine_counts = df['Cuisines'].value_counts()

# Step 2:The top 3 cuisines
top_cuisines = cuisine_counts.head(3)

# Step 3: Calculate the percentage of restaurants serving these cuisines
total_restaurants = len(df)
top_cuisines_percentage = (top_cuisines / total_restaurants) * 100

# Display the results
print("Top 3 Cuisines:")
print(top_cuisines)
print("\nPercentage of Restaurants Serving Top Cuisines:")
print(top_cuisines_percentage)

# Step 4: Visualize the results
plt.figure(figsize=(8, 5))
top_cuisines.plot(kind='bar', color='skyblue')
plt.title('Top 3 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

**TASK 2**

> **CITY ANALYSIS**

* Step 1: Count the number of restaurants in each city
* Calculate the average rating for restaurants in each city
* Find the city with the highest average rating
* Step 4: Visualize the top cities with the most restaurants







```python
#CITY ANALYSIS
# Step 1: Count the number of restaurants in each city
city_counts = df['City'].value_counts()

# City with the highest number of restaurants
most_restaurants_city = city_counts.idxmax()
most_restaurants_count = city_counts.max()

print(f"The city with the highest number of restaurants is {most_restaurants_city} with {most_restaurants_count} restaurants.")

# Step 2: Calculate the average rating for restaurants in each city
city_avg_ratings = df.groupby('City')['Votes'].mean()

# Step 3: Find the city with the highest average rating
highest_avg_rating_city = city_avg_ratings.idxmax()
highest_avg_rating = city_avg_ratings.max()

print(f"The city with the highest average rating is {highest_avg_rating_city} with an average rating of {highest_avg_rating:.2f}.")

# Step 4: Visualize the top cities with the most restaurants
plt.figure(figsize=(10, 6))
city_counts.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Cities with the Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

**TASK 3**


> **PRICE RANGE DISTRIBUTION**


* Step 1: Analyze the Price Range Distribution
* Step 2: Calculate the percentage of restaurants in each price range
* Step 3: Visualize the Price Range Distribution







```python
#PRICE RANGE DISTRIBUTION
# Step 1: Analyze the Price Range Distribution
price_range_counts = df['Price range'].value_counts()

# Step 2: Calculate the percentage of restaurants in each price range
total_restaurants = len(df)
price_range_percentage = (price_range_counts / total_restaurants) * 100

# Display the results
print("Price Range Distribution:")
print(price_range_counts)
print("\nPercentage of Restaurants in Each Price Range:")
print(price_range_percentage)

# Step 3: Visualize the Price Range Distribution
plt.figure(figsize=(8, 5))
price_range_counts.plot(kind='hist', color='teal')
plt.title('Price Range Distribution')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()

#Pie Chart for Percentage Distribution
plt.figure(figsize=(8, 8))
price_range_percentage.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Percentage of Restaurants by Price Range')
plt.ylabel('')
plt.show()
```

**TASK 4**

> **ONLINE DELIVERY**

* Calculate the percentage of restaurants that offer online delivery
* Compare Average Ratings



```python
#ONLINE DELIVERY ANALYSIS
# Step 1: Analyze Online Delivery Availabily
online_delivery_counts = df['Restaurant Name'].value_counts()

# Calculate the percentage of restaurants that offer online delivery
total_restaurants = len(df)
online_delivery_percentage = (online_delivery_counts / total_restaurants) * 100

# Display the results
print("online delivery Availability:")
print(online_delivery_counts)
print("\nPercentage of Restaurants Offering online delivery:")
print(online_delivery_percentage)

# Step 2: Compare Average Ratings
average_rating_online = df[df['Restaurant Name'] == 'Yes']['City'].mean()
average_rating_no_online = df[df['Restaurant Name'] == 'No']['City'].mean()
print(f"\nAverage Rating for Restaurants with Has online delivery: {average_rating_online:.2f}")
print(f"\nAverage Rating for Restaurants without online delivery: {average_rating_no_online:.2f}")
```

# ***LEVEL 2***


> **TASK 1 RESTAURANT RATINGS**





* Step 1: Analyze the distribution of aggregate ratings
* Step 2: Calculate the average number of votes
* Step 3: Visualize the rating distribution





```python
# Step 1: Analyze the distribution of aggregate ratings
rating_distribution = df['Cuisines'].value_counts().sort_index()

# Find the most common rating range
# Bin ratings into ranges (e.g., 0-1, 1-2, ..., 4-5)
rating_bins = pd.cut(df['Aggregate rating'], bins=[0, 1, 2, 3, 4, 5], include_lowest=True)
rating_range_distribution = rating_bins.value_counts().sort_index()

# Step 2: Calculate the average number of votes
average_votes = df['Votes'].mean()

# Display the results
print("Cuisines Distribution:")
print(rating_distribution)
print("\nCuisines Range Distribution:")
print(rating_range_distribution)
print(f"\nAverage Number of Votes: {average_votes:.2f}")

# Step 3: Visualize the rating distribution
plt.figure(figsize=(8, 5))
rating_range_distribution.plot(kind='bar', color='purple')
plt.title('Distribution of Cuisines by Range')
plt.xlabel('Rating Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()

# Optional: Visualize the actual ratings
plt.figure(figsize=(8, 5))
rating_distribution.plot(kind='bar', color='blue')
plt.title('Distribution of Individual Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

**TASK 2**


> **CUISINE COMBINATION ANALYSIS**
* Step 1: Analyze Cuisine Combinations
* Step 2: Analyze Ratings for Common Combinations
* Step 3: Visualize the Results







```python
# Step 1: Analyze Cuisine Combinations
# Assuming the dataset has a column named 'Cuisines' with comma-separated values
df['Cuisines'] = df['Cuisines'].fillna('')  # Handle missing values
df['Cuisines Combinations'] = df['Cuisines'].apply(lambda x: tuple(sorted(x.split(', '))))

# Count the most common combinations
cuisine_combination_counts = df['Cuisines Combinations'].value_counts()

# Get the top 10 most common combinations
top_combinations = cuisine_combination_counts.head(10)

print("Top 10 Most Common Cuisines Combinations:")
print(top_combinations)

# Step 2: Analyze Ratings for Common Combinations
# Calculate the average rating for each combination
average_rating_by_combination = df.groupby('Cuisines Combinations')['Votes'].mean()

# Filter for the top 10 combinations
top_combination_ratings = average_rating_by_combination[top_combinations.index]

print("\nAverage Ratings for Top 10 Cuisine Combinations:")
print(top_combination_ratings)

# Step 3: Visualize the Results
import matplotlib.pyplot as plt

# Bar chart for top cuisine combinations
plt.figure(figsize=(10, 6))
top_combinations.plot(kind='bar', color='orange')
plt.title('Top 10 Cuisine Combinations')
plt.xlabel('Cuisine Combination')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.show()

# Bar chart for average ratings of top combinations
plt.figure(figsize=(10, 6))
top_combination_ratings.plot(kind='bar', color='green')
plt.title('Average Ratings for Top 10 Cuisine Combinations')
plt.xlabel('Cuisine Combination')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**TASK 3**

> **GEOGRAPHIC MAP ANALYSIS**


*   Step 1: Verify Columns for Longitude and Latitude
*   Step 2: Filter Data for Valid Coordinates
*   Step 3: Create a Base Map
*   Step 4: Add Markers for Each Restaurant
*   Step 5: Save the Map









```python
!pip install folium
import folium
from folium.plugins import MarkerCluster # Import MarkerCluster
# Step 1: Verify Columns for Longitude and Latitude
# Assuming the dataset has 'Longitude' and 'Latitude' columns
if 'Longitude' not in df.columns or 'Latitude' not in df.columns:
    print("Dataset must have 'Longitude' and 'Latitude' columns for geographic analysis.")
else:
    # Step 2: Filter Data for Valid Coordinates
    df = df.dropna(subset=['Longitude', 'Latitude'])  # Remove rows with missing coordinates
    df = df[(df['Longitude'] != 0) & (df['Latitude'] != 0)]  # Remove invalid coordinates

    # Step 3: Create a Base Map
    # Center the map around the mean coordinates
    center_lat = df['Latitude'].mean()
    center_lon = df['Longitude'].mean()
    base_map = folium.Map(location=[center_lat, center_lon], zoom_start=12)

    # Step 4: Add Markers for Each Restaurant
    marker_cluster = MarkerCluster().add_to(base_map)

    for _, row in df.iterrows():
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"Name: {row.get('Restaurant Name', 'N/A')}<br>Rating: {row.get('Rating', 'N/A')}",
        ).add_to(marker_cluster)

    # Step 5: Save the Map
    map_file = "restaurant_locations.html"
    base_map.save(map_file)
    print(f"Map saved as {map_file}. Open this file in a browser to view the map.")
```

**TASK 4**

> **RESTAURANT CHAIN ANALYSIS**
* Step 1: Identify Restaurant Chains
* Step 2: Analyze Ratings for Chains





```python
# Step 1: Identify Restaurant Chains
restaurant_counts = df['Restaurant Name'].value_counts()

# Define a chain as having more than one location
restaurant_chains = restaurant_counts[restaurant_counts > 1]

print("Top Restaurant Chains:")
print(restaurant_chains)

# Step 2: Analyze Ratings for Chains
# Filter the dataset for restaurants in chains
chain_data = df[df['Restaurant Name'].isin(restaurant_chains.index)]

# Calculate average ratings for each chain
chain_ratings = chain_data.groupby('Restaurant Name')['Votes'].mean()

# Step 3: Analyze Popularity (Votes) for Chains
chain_votes = chain_data.groupby('Restaurant Name')['Votes'].mean()

# Combine the results into a single DataFrame
chain_analysis = pd.DataFrame({
    'Average Rating': chain_ratings,
    'Average Votes': chain_votes,
    'Number of Locations': restaurant_chains
}).sort_values(by='Number of Locations', ascending=False)

print("\nRestaurant Chains Analysis:")
print(chain_analysis)
```
