```python
#Example 1: Sales Data (Small Business Simulation)
#Dataset Creation
import pandas as pd

data = {
    "Product": ["Rice", "Beans", "Yam", "Garri", "Oil", "Sugar", "Salt", "Milk"],
    "Category": ["Food", "Food", "Food", "Food", "Grocery", "Grocery", "Grocery", "Grocery"],
    "Quantity": [120, 80, 45, 60, 30, 50, 75, 40],
    "Price": [500, 400, 800, 350, 1000, 600, 200, 700],
    "Sales": [60000, 32000, 36000, 21000, 30000, 30000, 15000, 28000]
}

sales_df = pd.DataFrame(data)
print(sales_df)
```


 Exercises
1. Calculate the total revenue and average price.
2. Add a new column for Profit = 20% of Sales.
3. Identify which category generated the highest revenue.
4. Create a bar chart showing each productâ€™s sales.


```python
#Solutions
import matplotlib.pyplot as plt

#Total and Average
total_revenue = sales_df["Sales"].sum()
average_price = sales_df["Price"].mean()
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Average Price: ${average_price:,.2f}")

#Profit column (20% of Sales)
sales_df["Profit"] = sales_df["Sales"] * 0.2

#Top category by Sales
top_category = sales_df.groupby("Category")["Sales"].sum().idxmax()
print(f"Top Category: {top_category}")

# Bar chart: Product Sales Performance
plt.figure(figsize=(10,6))
bars = plt.bar(sales_df["Product"], sales_df["Sales"], color="skyblue")

plt.title("Product Sales Performance")
plt.xlabel("Product")
plt.ylabel("Sales ($)")

# Annotate bars with sales value and percentage of total revenue
for bar in bars:
    height = bar.get_height()
    pct = (height / total_revenue) * 100
    plt.text(bar.get_x() + bar.get_width()/2, height + total_revenue*0.01, f"${height:,.0f}\n({pct:.1f}%)",
             ha='center', va='bottom', fontsize=9)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Example 2: Student Grades Dataset (Education Context)
 #Dataset Creation
import pandas as pd

grades = {
    "Name": ["Aisha", "David", "Zainab", "Tunde", "Fatima", "John", "Blessing", "Peter"],
    "Math": [85, 70, 92, 60, 74, 88, 95, 67],
    "English": [78, 82, 89, 65, 80, 75, 91, 70],
    "Science": [90, 68, 94, 55, 72, 84, 97, 64],
    "Attendance": [95, 88, 98, 76, 85, 90, 100, 82]
}

grades_df = pd.DataFrame(grades)
print(grades_df)
```

Exercises
1. Calculate each studentâ€™s average score across subjects.
2. Identify the top 3 students overall.
3. Find the correlation between Attendance and Science.
4. Visualize average subject scores with a bar chart.


```python
import matplotlib.pyplot as plt

#Calculate Average Score per Student
grades_df["Average"] = grades_df[["Math", "English", "Science"]].mean(axis=1)

#Top 3 Students
top3 = grades_df.nlargest(3, "Average")[["Name", "Average"]]
print("Top 3 Students:\n", top3)

#Correlation
corr = grades_df["Attendance"].corr(grades_df["Science"])
print("Correlation between Attendance and Science:", round(corr, 2))

#Visualization: Average Score per Student
plt.figure(figsize=(10,6))
bars = plt.bar(grades_df["Name"], grades_df["Average"], color="skyblue")

plt.title("Average Score per Student")
plt.xlabel("Student Name")
plt.ylabel("Average Score")

# Annotate each bar with the score
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f"{height:.1f}", ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Example 3: NGO Training Impact Data (Social Sector)
#Dataset Creation
import pandas as pd

ngo_data = {
    "Region": ["North", "South", "East", "West", "Central"],
    "Year": [2020, 2021, 2022, 2023, 2024],
    "Participants": [120, 180, 240, 310, 400],
    "Female": [60, 100, 140, 180, 250],
    "Male": [60, 80, 100, 130, 150]
}

ngo_df = pd.DataFrame(ngo_data)
print(ngo_df)
```

Exercises
1. Calculate total participants over all years.
2. Compute the growth rate from 2020 to 2024.
3. Determine the female participation percentage for each year.
4. Visualize yearly participant growth trend.


```python
import matplotlib.pyplot as plt

#Calculate Female and Male Percentages
ngo_df["Female_%"] = (ngo_df["Female"] / ngo_df["Participants"]) * 100
ngo_df["Male_%"] = 100 - ngo_df["Female_%"]

#Display percentages
print("Participation by Gender:")
print(ngo_df[["Year", "Female_%", "Male_%"]].round(2))

# Visualization: Stacked Bar Chart
plt.figure(figsize=(10,6))

plt.bar(ngo_df["Year"], ngo_df["Female_%"], label="Female", color="green")
plt.bar(ngo_df["Year"], ngo_df["Male_%"], bottom=ngo_df["Female_%"], label="Male", color="blue")

plt.xlabel("Year")
plt.ylabel("Percentage of Participants")
plt.title("Participants by Gender")
plt.xticks(ngo_df["Year"])
plt.ylim(0, 100)

# Annotate each bar with percentages
for i, row in ngo_df.iterrows():
    plt.text(row["Year"], row["Female_%"]/2, f'{row["Female_%"]:.1f}%', ha='center', va='center', color='white', fontweight='bold')
    plt.text(row["Year"], row["Female_%"] + row["Male_%"]/2, f'{row["Male_%"]:.1f}%', ha='center', va='center', color='white', fontweight='bold')

plt.legend(title="Gender")
plt.tight_layout()
plt.show()

```

```python
#Example 4: Employee Performance Dataset (Corporate Context)
#Dataset Creation
import pandas as pd

employees = {
    "Employee": ["Amaka", "John", "Musa", "Zainab", "Tayo", "Grace", "Yusuf", "Helen"],
    "Department": ["HR", "IT", "Finance", "IT", "Marketing", "HR", "Finance", "IT"],
    "Experience(Years)": [2, 5, 3, 4, 6, 1, 3, 7],
    "Projects_Handled": [3, 10, 6, 8, 12, 2, 5, 15],
    "Performance_Score": [70, 88, 80, 85, 90, 68, 75, 95]
}

emp_df = pd.DataFrame(employees)
print(emp_df)
```

Exercises
1. Find the average performance score per department.
2. Identify the employee with the highest number of projects.
3. Analyze correlation between Experience and Performance Score.
4. Visualize performance by department using a bar chart.


```python
import matplotlib.pyplot as plt

#Departmental Average Performance
dept_perf = emp_df.groupby("Department")["Performance_Score"].mean()
print("Average Departmental Performance:\n", dept_perf.round(2))

#Top Project Handler
top_emp = emp_df.loc[emp_df["Projects_Handled"].idxmax()]
print(f"\nTop Project Handler: {top_emp['Employee']} with {top_emp['Projects_Handled']} projects")

#Correlation
corr = emp_df["Experience(Years)"].corr(emp_df["Performance_Score"])
print(f"\nCorrelation between Experience and Performance Score: {corr:.2f}")

#Visualization: Average Departmental Performance
plt.figure(figsize=(10,6))
bars = plt.bar(dept_perf.index, dept_perf.values, color="skyblue")

plt.title("Average Departmental Performance")
plt.xlabel("Department")
plt.ylabel("Performance Score")

# Annotate each bar with value
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f"{height:.1f}", ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Example 5: Local Business Expense Tracker
#Dataset Creation
import pandas as pd

expenses = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Rent": [120000, 120000, 120000, 120000, 120000, 120000],
    "Utilities": [20000, 18000, 22000, 19000, 25000, 21000],
    "Marketing": [15000, 25000, 30000, 20000, 27000, 18000],
    "Supplies": [10000, 12000, 9000, 11000, 13000, 12500]
}

expenses_df = pd.DataFrame(expenses)
print(expenses_df)
```

Exercises
1. Calculate the total monthly expenses.
2. Determine which month had the highest total spending.
3. Visualize monthly spending trend using a line chart.
4. Create a summary DataFrame with total and average expenses.


```python
import matplotlib.pyplot as plt

#Total Monthly Expenses
expenses_df["Total_Expense"] = expenses_df[["Rent", "Utilities", "Marketing", "Supplies"]].sum(axis=1)

#Monthly Expense
expenses_df["Expense_%"] = (expenses_df["Total_Expense"] / expenses_df["Total_Expense"].sum()) * 100

print("Monthly Expenses with % of Total:")
print(expenses_df[["Month", "Total_Expense", "Expense_%"]].round(2))

#Highest Spending Month
max_month_idx = expenses_df["Total_Expense"].idxmax()
max_month = expenses_df.loc[max_month_idx, "Month"]
print(f"\nHighest Spending Month: {max_month}")

#Visualization
plt.figure(figsize=(10,6))
plt.plot(expenses_df["Month"], expenses_df["Total_Expense"], marker='o', color="skyblue", label="Total Expense")

# Highlight the highest spending month
plt.scatter(expenses_df.loc[max_month_idx, "Month"], expenses_df.loc[max_month_idx, "Total_Expense"],
            color="red", s=100, label=f"Highest: {max_month}")

# Annotate each point with % of total
for i, row in expenses_df.iterrows():
    plt.text(row["Month"], row["Total_Expense"] + expenses_df["Total_Expense"].max()*0.01,
             f"{row['Expense_%']:.1f}%", ha='center', fontsize=9)

plt.title("Monthly Business Expenses")
plt.xlabel("Month")
plt.ylabel("Total Expense (â‚¦)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

#Summary
summary = pd.DataFrame({
    "Metric": ["Total Expense", "Average Monthly Expense"],
    "Value": [expenses_df["Total_Expense"].sum(), expenses_df["Total_Expense"].mean()]
})
print("\nSummary:")
print(summary)
```
